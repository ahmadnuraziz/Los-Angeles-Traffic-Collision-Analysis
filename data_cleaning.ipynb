```python
import re
latitude = np.zeros(len(data))
longitude = np.zeros(len(data))
for i in range(len(data)):
    latitude[i] = float(data.loc[i,'Location'].split(':')[1].split(',')[0].replace(' ','').replace("'",''))
    longitude[i] = float(data.loc[i,'Location'].split(':')[-1].replace("'",'').replace(' ','').replace('}',''))
```

```python
data['Latitude'] = pd.DataFrame({'Latitude':latitude})
data['Longitude'] = pd.DataFrame({'Longitude':longitude})
```
